```python
def fit_and_calcuate_metrixs(models:dict,X_train,X_test,y_train,y_test):
    models_info = {}
    for name,model in models.items():
        print(name)
        model = model.fit(X_train,y_train)
        y_preds = model.predict(X_test)
        try:
            info = {
                'Accuracy':model.score(X_test,y_test),
                'F1 Score':f1_score(y_test,y_preds),
                'Precision':precision_score(y_test,y_preds),
                'Recall':recall_score(y_test,y_preds)
            }
            plot_confusion_matrix(model,X_test,y_test)
            plot_roc_curve(model,X_test,y_test)
            models_info[name] = info
        except:
            info = {
                'Accuracy':model.score(X_test,y_test),
                'MAE':mean_absolute_error(y_test,y_preds),
                'MSE':mean_squared_error(y_test,y_preds)
            }
            models_info[name] = info
        print('\n\n')
    return models_info
def calcuate_metrixs(models:dict,X_train,X_test,y_train,y_test):
    models_info = {}
    for name,model in models.items():
        print(name)
        y_preds = model.predict(X_test)
        info = {
            'Accuracy':model.score(X_test,y_test),
            'F1 Score':f1_score(y_test,y_preds),
            'Precision':precision_score(y_test,y_preds),
            'Recall':recall_score(y_test,y_preds)
        }
        plot_confusion_matrix(model,X_test,y_test)
        plot_roc_curve(model,X_test,y_test)
        models_info[name] = info
        print('\n\n')
    return models_info
```

## RandomSearchCV Check
the winner from this will get a GridSearchCV :)

```python
def calcuate_metrixs(models:dict,X_train,X_test,y_train,y_test):
    models_info = {}
    for name,model in models.items():
        print(name)
        y_preds = model.predict(X_test)
        info = {
            'Accuracy':model.score(X_test,y_test),
            'F1 Score':f1_score(y_test,y_preds),
            'Precision':precision_score(y_test,y_preds),
            'Recall':recall_score(y_test,y_preds)
        }
        plot_confusion_matrix(model,X_test,y_test)
        plot_roc_curve(model,X_test,y_test)
        models_info[name] = info
        print('\n\n')
    return models_info
```

```python
grid_rfc = {
    "n_estimators":[10,100,125,250,500,750,1000,1250,2500],
    "max_depth":[None,5,10,15,20,25],
    "max_features":["auto","sqrt",'log2'],
    "min_samples_split":[2,4,5],
    "min_samples_leaf":[2,4,5],
    'bootstrap':[False,True],
    'warm_start':[False,True]
}
grid_dtc = {
    'criterion':['gini','entropy'],
    'splitter':['best','random'],
    'max_depth':[None,5,10,15,20,25],
    'min_samples_split':[2,4,5],
    'min_samples_leaf':[2,4,5],
    'max_features':["auto","sqrt",'log2'],
    'min_impurity_decrease':[2,4,5]
}
model_dtc_rs = RandomizedSearchCV(DecisionTreeClassifier(),param_distributions=grid_dtc,n_iter=50,cv=5,verbose=True).fit(X_train,y_train)
```

```python
model_rfc_rs = RandomizedSearchCV(RandomForestClassifier(),param_distributions=grid_rfc,n_iter=50,cv=5,verbose=25).fit(X_train,y_train)
```

```python
clf_results = calcuate_metrixs({'DTC':model_dtc_rs,'RFC':model_rfc_rs},X_train,X_test,y_train,y_test)
```

```python
models_clf_results_2 = pd.DataFrame(clf_results.values(),clf_results.keys())
```

```python
models_clf_results_2.plot.bar(figsize=(20,10))
```

## GridSearchCV
### RTC is the winner !!! *aka Random Forrest Classifier*

```python
grid_rfc = {
    "n_estimators":[100,2500,5000,10000],
    "max_depth":[None,25,50],
    "max_features":["auto","sqrt",'log2'],
    "min_samples_split":[5,10],
    "min_samples_leaf":[5,10],
    'bootstrap':[False,True],
    'warm_start':[False,True]
}
model_rfc_gs = GridSearchCV(RandomForestClassifier(),param_grid=grid_rfc,cv=round(2.5),verbose=25).fit(X_train,y_train)
```

```python
clf_results = calcuate_metrixs({'RFC':model_rfc_gs},X_train,X_test,y_train,y_test)
models_clf_results_3 = pd.DataFrame(clf_results.values(),clf_results.keys())
models_clf_results_3.plot.bar(figsize=(20,10))
```
